### Matplotlib
- it is a cross-platform data visualization library for making 2D plots from data

### Types of Plotting
- Line Plot
- Bar Graph
- Scatter Plot
- Area Plot 
- Pie Charts

```python
import matplotlib.pyplot as plt
```

### Lines

```python
plt.plot([1,2,3,4],[2,4,6,8])
plt.show()
```

```python
plt.plot([1,2,3,4],[2,4,6,8],'g')
plt.show()
```

### label(), title(), legend()

```python
import numpy as np
x = np.linspace(0,2,100)
x
```

```python
plt.plot(x,x,color = 'r',label = 'Normal Line')
plt.plot(x,x**2,color = 'g',label = 'Square Line')
plt.xlabel('X-Axes')
plt.ylabel('Y-Axes')
plt.title("Line Plotting")
plt.legend()
plt.show()
```

### Bar Graph
- Used to create a Bar Chart

```python
plt.bar([1,2,3,4],[20,30,40,50],width=0.4)
plt.show()
```

### Histgram
- Used for frequency distribution of numerical data

```python
a = [0,1,1,1.2,1.3,2.6,2.7,3,5,5.9,6]
bins = 5
plt.hist(a,bins, width=0.8)
```

```python
a2 = [0,5,8,6,13,16,16.9,18,25,29,33,36,37,60]
bins = [0,10,22,30,42,50,62]
plt.hist(a2,bins,rwidth=0.8)
```

### Box Plot: Used for displaying the distribution of data base on min, IQR, max values
- min
- IQR(Inter Quartile Range)
    - Lower Quartile(25%)
    - Median(50%)
    - Upper Quartile(75%)
- max

```python
marks = np.random.randint(20,100,101)
marks
```

```python
plt.boxplot(marks,patch_artist=True)
plt.show()
```

```python
v1 = np.random.randint(50,100,10)
v2 = np.random.randint(40,90,10)
v3 = np.random.randint(30,100,10)
plt.boxplot([v1,v2,v3],labels = ['val1','val2','val3'])
plt.show()
```

### Scatter Plot

plt.scatter(): similar to line plotting.

```python
plt.scatter([1,2,3,4],[2,4,6,8])
plt.show()
```

### Area Plot
- plt.stackplot()

```python
days = [1,2,3,4]
working = [8,11,9,10]
sleeping = [8,10,11,12]

plt.stackplot(days,working,sleeping,colors = ['r','g'],labels = ['Work','Rest'])
plt.legend()
plt.show()
```
