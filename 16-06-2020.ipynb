## Pandas:

* Pandas is a libraray in python which is used to reduce and make our work siple and esay.
* Pandas can be applicable for any kind of data(ordered,unordered,stastical data etc)
* Pandas represent output in the format of table.

#### Main keywords in creating data:

* Series -- In this series format output will be repesented row by row(line by lne)
* DataFrame -- In this data will be represented in the format row and columns.


```python
import pandas as pd
import numpy as np
```

```python
#### Creating data with series:
## pd.Series([list])

a = pd.Series([1,2,3,4,5,6])
a
```

```python
### Creating Data Series using Numpy array:

a = np.array([1,2,3,4,5,6])
a
```

```python
d = pd.Series(a)
d
```

```python
### Creating Data Series using Dictionary:

d = {'a':0,'b':2,"c":4}
d
```

```python
a = pd.Series(d)
a
```

```python
### Changing the index values in Series:

a = pd.Series([1,2,3,4,5,6],index = [1,2,3,4,5,6])
a
```

```python
a = pd.Series([2,3,4,5,6],index = np.arange(1,6))
a
```

```python
a.dtype
```

```python
b = pd.Series(['a','b','c'])
b
```

```python
b.dtype
```

```python
#### Creating data using DataFrames:

a = pd.DataFrame([1,2,3,4,5])
a
```

```python
### Creating multiple columns in a Dataframe:

a = [["Niharikaa",12],["Lavanaya",10],["Sireesha",15],["Mounica",20]]
data = pd.DataFrame(a)
data
```

```python
### By using list creating multiple columns:

a = [["Niharikaa",12],["Lavanaya",10],["Sireesha",15],["Mounica",20]]
data = pd.DataFrame(a,columns = ("Name","Age"))
data
```

```python
### Creating multiple columns values using Dictionary:

d = {"Name":["Niharikaa","Lavanaya","Mounica","sireesha","Sudha"],
    "Age":[12,14,16,20,30],
    "Branch":["ECE","CSE","ECE","MECH","CIVIL"]}
data = pd.DataFrame(d,index = np.arange(1,6))
data
```

```python
d = {"name":["niha","sanju","mouni","mounica","sravani"],
    "Age":[12,14,16,20,30],
    "Branch":["ece","cse","mech","eee","civil"]}
data = pd.DataFrame(d,index = np.arange(1,6))
data
```

```python
a = pd.DataFrame([1,2,3,4,5],index = ["A","B","C","D","E"])
a
```

```python
#### Creating the Dataframe from multile dictionaries:

data = [{'a':1,'b':2,'c':3},{'a':4,'b':5,'c':7}]
d = pd.DataFrame(data)
d
```

```python
#### converting data series into dataframe:
## Not a  number (NaN)
d = {"one" : pd.Series([1,2,3],index = ['a','b','c']),
     "two":pd.Series([1,2,3,4],index = ['a','b','c','d'])}
data = pd.DataFrame(d)
data
```

```python
a = [1,2,3,4]
a = pd.DataFrame(a,dtype = 'float64')
a
```

```python
d = {"one" : pd.Series([1,2,3],index = ['a','b','c']),
     "two":pd.Series([1,2,3,4],index = ['a','b','c','d'])}
data = pd.DataFrame(d)
data
```
